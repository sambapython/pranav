```python
#pickle
import json
path = os.path.join("data","data.json")
f=open(path, "w")
json.dump(data,f)
f.close()
f=open(path, "r")
d=json.load(f)
print(d, type(d))
f.close()
```

```python
#pickle
import yaml
from yaml.loader import SafeLoader
path = os.path.join("data","data.yaml")
f=open(path, "w")
yaml.dump(data,f)
f.close()
f=open(path, "r")
d=yaml.load(f, Loader=SafeLoader)
print(d, type(d))
f.close()
```
